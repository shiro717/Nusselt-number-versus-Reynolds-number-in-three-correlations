```python
import numpy as np
import matplotlib.pyplot as plt

def hilpert_nu(Re, Pr):
    hilpert_params = [
        (0.4, 4, 0.989, 0.33),
        (4, 40, 0.911, 0.385),
        (40, 4000, 0.683, 0.466),
        (4000, 40000, 0.193, 0.618),
        (40000, 400000, 0.027, 0.805),
    ]
    for Re_min, Re_max, C, m in hilpert_params:
        if Re_min <= Re < Re_max:
            return C * Re**m * Pr**(1/3)
    return np.nan

def zukauskas_nu(Re, Pr, Pr_s):
    zukauskas_params = [
        (1, 40, 0.75, 0.4),
        (40, 1000, 0.51, 0.5),
        (1000, 2e5, 0.26, 0.6),
        (2e5, 1e6, 0.076, 0.7),
    ]
    n = 0.36 if Pr > 10 else 0.37
    for Re_min, Re_max, C, m in zukauskas_params:
        if Re_min <= Re < Re_max:
            return C * Re**m * Pr**n * (Pr / Pr_s)**(1/4)
    return np.nan

def churchill_bernstein_nu(Re, Pr):
    return (
        0.3 +
        0.62 * Re**0.5 * Pr**(1/3) /
        (1 + (0.4 / Pr)**(2/3))**(1/4) *
        (1 + (Re / 282000)**(5/8))**(4/5)
    )

Re_range = np.logspace(-0.4, 6, 500)
Pr = 0.684
Pr_s = 0.684

nu_hilpert = [hilpert_nu(Re, Pr) for Re in Re_range]
nu_zukauskas = [zukauskas_nu(Re, Pr, Pr_s) for Re in Re_range]
nu_churchill = [churchill_bernstein_nu(Re, Pr) for Re in Re_range]

plt.figure(figsize=(10, 6))
plt.loglog(Re_range, nu_hilpert, label='Hilpert', linewidth=2)
plt.loglog(Re_range, nu_zukauskas, label='Zukauskas', linewidth=2)
plt.loglog(Re_range, nu_churchill, label='Churchill-Bernstein', linewidth=2)

plt.xlabel('Reynolds Number (Re)', fontsize=14)
plt.ylabel('Nusselt Number (Nu)', fontsize=14)
plt.title('Nusselt Number vs. Reynolds Number', fontsize=16)
plt.legend()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
